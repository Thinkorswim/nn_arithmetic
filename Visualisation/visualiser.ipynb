# Activation functions

```python
def tanh(x):
    return (2 / (1 + math.exp(-2*x))) - 1

def relu(x):
    return max(0,x)

def selu(x):
    lamb = 1.0507
    alpha = 1.67326
    
    if x>0:
        return lamb*x
    else:
        return lamb*(alpha*(math.exp(x) - 1))
    
def softsign(x):
    return x / (1 + abs(x))
        
```
